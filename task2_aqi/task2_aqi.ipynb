# Introduction
Delhi faces severe air pollution challenges due to vehicle emissions, industrial activity, crop stubble burning, and geographical factors.  

This study analyzes hourly AQI data to identify pollutant patterns, seasonal variations, and correlations.

---

# Research Questions
1. Which pollutants contribute most to air pollution in Delhi?  
2. How do pollutant levels vary by season/month?  
3. Which pollutants are highly correlated?  
4. What recommendations can improve air quality and public health?

---

# Findings
- **Descriptive Stats →** PM2.5 and PM10 often exceed safe limits.  
- **Daily/Monthly Trends →** Winter months (Dec–Jan) show the worst pollution.  
- **Seasonal Variation →** PM2.5 and PM10 are highest in winter; O₃ higher in summer.  
- **Correlation →** PM2.5 strongly correlates with PM10; gaseous pollutants (NO₂, SO₂) show different patterns.  

---

# Conclusion & Recommendations
- **Winter Pollution Spike:** Likely caused by stubble burning + temperature inversion trapping pollutants.  
- **Vehicular Emissions:** Strong contributor to NO₂; stricter emission standards required.  
- **Industrial/Construction Dust:** Source of PM10 → dust control measures needed.  
- **Targeted Policies:** Odd-even vehicle schemes, promoting public transport, stricter crop residue burning enforcement, and investment in renewable energy.
